```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import imdb
from tornasole.trials import LocalTrial
```

```python
lt = LocalTrial( 'sentiment', 'ts_output/', parallel=True)
```

```python
lt.tensors()
```

```python
def tplt( trial, tname ): 
    t = lt.tensor(tname)
    steps = t.steps()
    _t = [t.value(s) for s in steps]
    plt.plot( steps, _t, label=tname)
```

```python
tplt( lt, 'acc')
tplt( lt, 'val_acc')
plt.legend()
plt.show()
```

```python
tplt( lt, 'loss')
tplt( lt, 'val_loss')
plt.legend()
plt.show()
```

```python
tplt( lt, 'mean_squared_error')
tplt( lt, 'val_mean_squared_error')
plt.legend()
plt.show()
```
